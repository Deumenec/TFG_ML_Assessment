```python
import math
import numpy as np
import matplotlib.pyplot as plt
from Bio.PDB import PDBParser, PPBuilder
from Bio.PDB.vectors import calc_dihedral
import os

# --- Function to Extract Omega Angles (No change) ---
def extract_omega_angles(file_name, data_directory="dades"):
    """
    Extracts omega angles from a PDB file.

    Args:
        file_name (str): The base name of the PDB file (e.g., "3qbi").
        data_directory (str): The directory where the PDB files are located.

    Returns:
        list: A list of omega angles in degrees.
    """
    pdb_file_path = os.path.join(data_directory, f"{file_name}.pdb")
    
    try:
        parser = PDBParser(QUIET=True)
        structure = parser.get_structure("protein", pdb_file_path)
    except FileNotFoundError:
        print(f"Error: The file was not found at {pdb_file_path}")
        return []

    omega_angles = []
    ppb = PPBuilder()

    print(f"Extracting omega angles from {file_name}...")
    for model in structure:
        for chain in model:
            peptides = ppb.build_peptides(chain)
            for poly in peptides:
                for i in range(len(poly) - 1):
                    res1 = poly[i]      # Residue i-1
                    res2 = poly[i+1]    # Residue i
                    
                    try:
                        # Atoms for omega: CA(i-1) - C(i-1) - N(i) - CA(i)
                        ca_prev_vector = res1["CA"].get_vector()
                        c_prev_vector = res1["C"].get_vector()
                        n_curr_vector = res2["N"].get_vector()
                        ca_curr_vector = res2["CA"].get_vector()

                        omega = calc_dihedral(ca_prev_vector, c_prev_vector, n_curr_vector, ca_curr_vector)
                        omega_angles.append(math.degrees(omega))
                    except KeyError:
                        pass # Skip if atoms are missing
    
    return omega_angles

# --- Function to Plot Omega Angle Distributions (Histogram) ---
def plot_omega_distributions(omega_data_dict, title_suffix="", output_directory="omegas"):
    """
    Plots overlaid histograms of omega angle distributions for multiple datasets.

    Args:
        omega_data_dict (dict): A dictionary where keys are labels (e.g., "3qbi")
                                and values are lists of omega angles.
        title_suffix (str): An additional string to add to the plot title.
        output_directory (str): Directory to save the plot.
    """
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # --- Set a global font family (optional) ---
    plt.rcParams["font.family"] = "Helvetica" 
    
    fig, ax1 = plt.subplots(figsize=(5, 4)) # Slightly larger figure size for better spacing
    
    # Activate grid
    ax1.grid(False) 

    # Centered range for plotting trans at 180 and cis at 0
    hist_range = (150, 210) 
    # Define bins for the histogram
    bins = np.arange(hist_range[0], hist_range[1] + 10, 1.25)

    colors = ['#daa520', '#3cb371']
    label_bo = ["Experimental Structure", "Predicted Structure"]
    
    # Store histogram data to pass to ax1.hist
    all_data_for_hist = []
    labels_for_hist = []

    for i, (label, omega_angles) in enumerate(omega_data_dict.items()):
        if not omega_angles:
            print(f"No omega angles for {label}. Skipping plot for this dataset.")
            continue

        # Angle Transformation
        adjusted_omega_angles = [angle + 360 if angle < -90 else angle for angle in omega_angles]
        
        all_data_for_hist.append(adjusted_omega_angles)
        labels_for_hist.append(label_bo[i])

    # Plot the histograms
    ax1.hist(all_data_for_hist, bins=bins, color=colors, alpha=0.7,
             label=labels_for_hist, histtype="stepfilled", edgecolor='black', linewidth=0.8, density = True)

    # --- Aesthetics and Labels with Increased Font Size ---
    #ax1.set_title(f"Omega Angle Distribution\n({title_suffix})", fontsize=20, pad=10) # Added a title with larger font
    ax1.set_xlabel("Omega (ω) Angle (°)", fontsize=19)  # Changed from 14 to 18
    ax1.set_ylabel("Angle Frequency", fontsize=19)    # Changed from 14 to 18
    
    # Set X-axis limits and ticks
    ax1.set_xlim(hist_range)
    ax1.set_xticks(np.arange(hist_range[0], hist_range[1] + 1, 10))

    # --- Vertical Lines for Cis/Trans ---
    ax1.axvline(180, color='gray', linestyle='--', linewidth=2, label='Trans conformation (180°)')
    
    # --- Increase Tick Label Size ---
    ax1.tick_params(axis='both', which='major', labelsize=19) # Changed from 14 to 16
    
    # --- Increase Legend Font Size ---
    # Legend placement
    fig.legend(loc="upper right", bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes, fontsize=19) # Changed from 12 to 16
    
    plt.tight_layout() # Adjust plot to prevent labels from overlapping

    # --- Save Plot ---
    sanitized_suffix = title_suffix.replace(' ', '_').replace('(', '').replace(')', '').replace('/', '_')
    output_pdf_path = os.path.join(output_directory, f"omega_histogram_comparison_plot_{sanitized_suffix}.pdf")
    plt.savefig(output_pdf_path)
    print(f"Comparison plot saved to {output_pdf_path}")
    plt.show()
    plt.close() # Close the plot to free memory

# --- Main Execution Block (No change) ---
if __name__ == "__main__":
    # Define your protein file names
    main_protein_name = "7unl" 
    protein1_name = main_protein_name 
    protein2_name = main_protein_name + "_pred" 

    # Extract omega angles for each protein
    omega_angles_prot1 = extract_omega_angles(protein1_name)
    omega_angles_prot2 = extract_omega_angles(protein2_name)

    # --- Prepare directory and file for saving statistics (in 'omegas' folder) ---
    output_directory = "omegas"
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    stats_file_name = f"omega_stats_{protein1_name}_vs_{protein2_name}.txt"
    stats_file_path = os.path.join(output_directory, stats_file_name)

    # --- Calculate and save statistics for each protein ---
    print("\n--- Saving Omega Angle Statistics ---")
    with open(stats_file_path, 'w') as f:
        # For protein 1
        if omega_angles_prot1:
            adjusted_omega_angles_prot1 = [angle + 360 if angle < -90 else angle for angle in omega_angles_prot1]
            mean_prot1 = np.mean(adjusted_omega_angles_prot1)
            std_prot1 = np.std(adjusted_omega_angles_prot1)
            
            f.write(f"{protein1_name}:\n")
            f.write(f"Mean Omega Angle: {mean_prot1:.2f} degrees\n")
            f.write(f"Standard Deviation: {std_prot1:.2f} degrees\n\n")
        else:
            f.write(f"No angles for {protein1_name} to calculate statistics.\n\n")

        # For protein 2
        if omega_angles_prot2:
            adjusted_omega_angles_prot2 = [angle + 360 if angle < -90 else angle for angle in omega_angles_prot2]
            mean_prot2 = np.mean(adjusted_omega_angles_prot2)
            std_prot2 = np.std(adjusted_omega_angles_prot2)
            
            f.write(f"{protein2_name}:\n")
            f.write(f"  Mean Omega Angle: {mean_prot2:.2f} degrees\n")
            f.write(f"  Standard Deviation: {std_prot2:.2f} degrees\n")
        else:
            f.write(f"No angles for {protein2_name} to calculate statistics.\n")
    
    print(f"Omega angle statistics saved to: {stats_file_path}")

    # Store angles in a dictionary for easy plotting
    all_omega_data = {
        protein1_name: omega_angles_prot1,
        protein2_name: omega_angles_prot2
    }
    print("Cis residues experimental")
    for i in range(len( all_omega_data[protein1_name])):
        if (all_omega_data[protein1_name][i]<90 and all_omega_data[protein1_name][i]>-30):
            print(i, all_omega_data[protein1_name][i])
    print("Cis residues predicted")
    for i in range(len( all_omega_data[protein2_name])):
        if (all_omega_data[protein2_name][i]<90 and all_omega_data[protein2_name][i]>-30):
            print(i, all_omega_data[protein2_name][i])
    # Plot the distributions for comparison
    plot_omega_distributions(all_omega_data, title_suffix=f"{protein1_name} vs {protein2_name}", output_directory=output_directory)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from Bio.PDB import PDBParser, PPBuilder
from Bio.PDB.vectors import calc_dihedral
from scipy.stats import gaussian_kde
import os

# --- Function to Extract Omega Angles (No change) ---
def extract_omega_angles(file_name, data_directory="dades"):
    """
    Extracts omega angles from a PDB file.

    Args:
        file_name (str): The base name of the PDB file (e.g., "3qbi").
        data_directory (str): The directory where the PDB files are located.

    Returns:
        list: A list of omega angles in degrees.
    """
    pdb_file_path = os.path.join(data_directory, f"{file_name}.pdb")
    
    try:
        parser = PDBParser(QUIET=True)
        structure = parser.get_structure("protein", pdb_file_path)
    except FileNotFoundError:
        print(f"Error: The file was not found at {pdb_file_path}")
        return []

    omega_angles = []
    ppb = PPBuilder()

    print(f"Extracting omega angles from {file_name}...")
    for model in structure:
        for chain in model:
            peptides = ppb.build_peptides(chain)
            for poly in peptides:
                for i in range(len(poly) - 1):
                    res1 = poly[i]      # Residue i-1
                    res2 = poly[i+1]    # Residue i
                    
                    try:
                        # Atoms for omega: CA(i-1) - C(i-1) - N(i) - CA(i)
                        ca_prev_vector = res1["CA"].get_vector()
                        c_prev_vector = res1["C"].get_vector()
                        n_curr_vector = res2["N"].get_vector()
                        ca_curr_vector = res2["CA"].get_vector()

                        omega = calc_dihedral(ca_prev_vector, c_prev_vector, n_curr_vector, ca_curr_vector)
                        omega_angles.append(math.degrees(omega))
                    except KeyError:
                        pass # Skip if atoms are missing
    
    return omega_angles

# --- Function to Plot Omega Angle Distributions (KDEs) - STYLED ---
def plot_omega_distributions(omega_data_dict, title_suffix="", output_directory="omegas"):
    """
    Plots overlaid KDEs of omega angle distributions for multiple datasets,
    with a style similar to the provided triple_ploter.

    Args:
        omega_data_dict (dict): A dictionary where keys are labels (e.g., "3qbi")
                                 and values are lists of omega angles.
        title_suffix (str): An additional string to add to the plot title.
        output_directory (str): Directory to save the plot.
    """
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Use a figure and a single axes object for KDEs (no twinx needed for density comparison)
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Activate grid similar to triple_ploter's commented-out line
    ax1.grid(True, which="both", linestyle="--", linewidth=0.4) 

    # Centered range for plotting trans at 180 and cis at 0
    hist_range = (140, 220) 
    x_kde = np.linspace(hist_range[0], hist_range[1], 500) # x-values for KDE plots

    # Define a list of distinct colors, in the colors I choose 
    colors = ['#daa520', '#3cb371']
    label_bo = ["Experimental structure", "Predicted structure"]
    for i, (label, omega_angles) in enumerate(omega_data_dict.items()):
        if not omega_angles:
            print(f"No omega angles for {label}. Skipping plot for this dataset.")
            continue

        # Angle Transformation: Shift angles from [-180, -90) to [180, 270)
        # to center the trans peak and align cis at 0.
        adjusted_omega_angles = [angle + 360 if angle < -90 else angle for angle in omega_angles]

        if len(adjusted_omega_angles) > 1: # KDE requires more than one point
            kde = gaussian_kde(adjusted_omega_angles)
            y_kde = kde(x_kde)
            
            # Plot the raw density for comparison
            # Using linewidth=2.5 and alpha=0.9 for lines, inspired by bar plot width/alpha
            ax1.plot(x_kde, y_kde, color=colors[i % len(colors)], linewidth=2.5, alpha=0.9, linestyle='-', label=f'{label_bo[i]}')
        else:
            print(f"Not enough data points for KDE for {label}.")

    # --- Aesthetics and Labels ---
    ax1.set_xlabel("Omega (ω) Angle [degrees]", fontsize=12)
    ax1.set_ylabel("Density", fontsize=12) # Y-axis represents probability density
    
    # Title format similar to triple_ploter
    ax1.set_title(f"Distribution of Omega Angles for {title_suffix}", fontsize=14, weight='bold')
    
    # --- Set X-axis limits and ticks ---
    ax1.set_xlim(hist_range)
    ax1.set_xticks(np.arange(hist_range[0], hist_range[1] + 1, 20))

    # --- Vertical Lines for Cis/Trans (subtler for comparison plots) ---
    ax1.axvline(180, color='gray', linestyle='--', linewidth=1.5, label='Trans conformation(180°)')
    
    # Legend placement similar to triple_ploter
    fig.legend(loc="upper right", bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)
    
    plt.tight_layout() # Adjust plot to prevent labels from overlapping
    
    # --- Save Plot ---
    # Sanitize title_suffix for filename to avoid issues with special characters
    sanitized_suffix = title_suffix.replace(' ', '_').replace('(', '').replace(')', '').replace('/', '_')
    output_pdf_path = os.path.join(output_directory, f"omega_comparison_plot_{sanitized_suffix}.pdf")
    plt.savefig(output_pdf_path)
    print(f"Comparison plot saved to {output_pdf_path}")
    plt.show()
    plt.close() # Close the plot to free memory

# Main
if __name__ == "__main__":
    # Define your protein file names
    main_protein_name = "9cah" 
    protein1_name = main_protein_name
    protein2_name = main_protein_name + "_pred" 

    # Extract omega angles for each protein
    omega_angles_prot1 = extract_omega_angles(protein1_name)
    omega_angles_prot2 = extract_omega_angles(protein2_name)

    # --- Prepare directory and file for saving statistics (in 'omegas' folder) ---
    output_directory = "omegas"
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    stats_file_name = f"omega_stats_{protein1_name}_vs_{protein2_name}.txt"
    stats_file_path = os.path.join(output_directory, stats_file_name)

    # --- Calculate and save statistics for each protein ---
    print("\n--- Saving Omega Angle Statistics ---")
    with open(stats_file_path, 'w') as f:
        f.write("--- Omega Angle Statistics ---\n\n")

        # For protein 1
        if omega_angles_prot1:
            adjusted_omega_angles_prot1 = [angle + 360 if angle < -90 else angle for angle in omega_angles_prot1]
            mean_prot1 = np.mean(adjusted_omega_angles_prot1)
            std_prot1 = np.std(adjusted_omega_angles_prot1)
            
            f.write(f"{protein1_name}:\n")
            f.write(f"  Mean Omega Angle: {mean_prot1:.2f} degrees\n")
            f.write(f"  Standard Deviation: {std_prot1:.2f} degrees\n\n")
        else:
            f.write(f"No angles for {protein1_name} to calculate statistics.\n\n")

        # For protein 2
        if omega_angles_prot2:
            adjusted_omega_angles_prot2 = [angle + 360 if angle < -90 else angle for angle in omega_angles_prot2]
            mean_prot2 = np.mean(adjusted_omega_angles_prot2)
            std_prot2 = np.std(adjusted_omega_angles_prot2)
            
            f.write(f"{protein2_name}:\n")
            f.write(f"  Mean Omega Angle: {mean_prot2:.2f} degrees\n")
            f.write(f"  Standard Deviation: {std_prot2:.2f} degrees\n")
        else:
            f.write(f"No angles for {protein2_name} to calculate statistics.\n")
    
    print(f"Omega angle statistics saved to: {stats_file_path}")

    # Store angles in a dictionary for easy plotting
    all_omega_data = {
        protein1_name: omega_angles_prot1,
        protein2_name: omega_angles_prot2
    }

    # Plot the distributions for comparison
    plot_omega_distributions(all_omega_data, title_suffix=f"{protein1_name} vs {protein2_name}", output_directory=output_directory)
```

```python
#Local chirality check
import MDAnalysis as mda
import numpy as np
import warnings

def check_chirality(pdb_file):
    """
    Analyzes the chirality of each amino acid in a protein by calculating
    the improper dihedral angle of the alpha-carbon.

    This function iterates through all residues in the provided PDB file.
    For each residue, it identifies the C, CA, N, and CB atoms to define
    the improper dihedral angle. Based on the sign of this angle, it
    classifies the amino acid as L-chiral, D-chiral, or achiral (in the
    case of Glycine).

    Args:
        pdb_file (str): The path to the PDB file to be analyzed.

    Returns:
        None: This function prints the chirality analysis directly to the console.
    """
    try:
        universe = mda.Universe(pdb_file)
    except Exception as e:
        print(f"Error loading PDB file: {e}")
        return

    protein = universe.select_atoms("protein")

    print(f"Analyzing chirality for {pdb_file}...")
    print("-" * 40)

    for residue in protein.residues:
        # Glycine is achiral, so we'll handle it as a special case.
        if residue.resname == "GLY":
            #print(f"Residue {residue.resname}{residue.resid}: Achiral")
            continue

        try:
            # Atom selections for the improper dihedral calculation
            c_atom = residue.atoms.select_atoms("name C")
            ca_atom = residue.atoms.select_atoms("name CA")
            n_atom = residue.atoms.select_atoms("name N")
            cb_atom = residue.atoms.select_atoms("name CB")

            # Ensure all necessary atoms are present
            if not all([len(group) == 1 for group in [c_atom, ca_atom, n_atom, cb_atom]]):
                warnings.warn(f"Could not find all atoms for residue {residue.resname}{residue.resid}. Skipping.")
                continue

            # Calculate the improper dihedral angle
            improper_dihedral = mda.lib.distances.calc_dihedrals(
                c_atom.positions[0],
                ca_atom.positions[0],
                n_atom.positions[0],
                cb_atom.positions[0]
            )
            
            # Convert from radians to degrees
            improper_dihedral_deg = np.rad2deg(improper_dihedral)

            # Determine chirality based on the sign of the dihedral angle
            if improper_dihedral_deg > 0:
                chirality = "L-chiral"
            else:
                chirality = "D-chiral"
                print(f"Residue {residue.resname}{residue.resid}: {chirality} (Improper Dihedral: {improper_dihedral_deg:.2f}°)")

        except Exception as e:
            warnings.warn(f"An error occurred for residue {residue.resname}{residue.resid}: {e}")

if __name__ == '__main__':
    # You will need to create a dummy PDB file for this example to run.
    # For instance, you can create a file named "protein.pdb" with
    # a simple amino acid structure.
    
    # Create a dummy PDB file for demonstration purposes        
    check_chirality("dades/longchain_pred.pdb")
    print("-" * 40)
    print("Analysis done!")
```

```python
#Not used KDE plot
import math
import numpy as np
import matplotlib.pyplot as plt
from Bio.PDB import PDBParser, PPBuilder
from Bio.PDB.vectors import calc_dihedral
from scipy.stats import gaussian_kde
import os

# --- Function to Extract Omega Angles (No change) ---
def extract_omega_angles(file_name, data_directory="dades"):
    """
    Extracts omega angles from a PDB file.

    Args:
        file_name (str): The base name of the PDB file (e.g., "3qbi").
        data_directory (str): The directory where the PDB files are located.

    Returns:
        list: A list of omega angles in degrees.
    """
    pdb_file_path = os.path.join(data_directory, f"{file_name}.pdb")
    
    try:
        parser = PDBParser(QUIET=True)
        structure = parser.get_structure("protein", pdb_file_path)
    except FileNotFoundError:
        print(f"Error: The file was not found at {pdb_file_path}")
        return []

    omega_angles = []
    ppb = PPBuilder()

    print(f"Extracting omega angles from {file_name}...")
    for model in structure:
        for chain in model:
            peptides = ppb.build_peptides(chain)
            for poly in peptides:
                for i in range(len(poly) - 1):
                    res1 = poly[i]      # Residue i-1
                    res2 = poly[i+1]    # Residue i
                    
                    try:
                        # Atoms for omega: CA(i-1) - C(i-1) - N(i) - CA(i)
                        ca_prev_vector = res1["CA"].get_vector()
                        c_prev_vector = res1["C"].get_vector()
                        n_curr_vector = res2["N"].get_vector()
                        ca_curr_vector = res2["CA"].get_vector()

                        omega = calc_dihedral(ca_prev_vector, c_prev_vector, n_curr_vector, ca_curr_vector)
                        omega_angles.append(math.degrees(omega))
                    except KeyError:
                        pass # Skip if atoms are missing
    
    return omega_angles

# --- Function to Plot Omega Angle Distributions (KDEs) - STYLED ---
def plot_omega_distributions(omega_data_dict, title_suffix="", output_directory="omegas"):
    """
    Plots overlaid KDEs of omega angle distributions for multiple datasets,
    with a style similar to the provided triple_ploter.

    Args:
        omega_data_dict (dict): A dictionary where keys are labels (e.g., "3qbi")
                                 and values are lists of omega angles.
        title_suffix (str): An additional string to add to the plot title.
        output_directory (str): Directory to save the plot.
    """
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Use a figure and a single axes object for KDEs (no twinx needed for density comparison)
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Activate grid similar to triple_ploter's commented-out line
    ax1.grid(False)  

    # Centered range for plotting trans at 180 and cis at 0
    hist_range = (120, 240) 
    x_kde = np.linspace(hist_range[0], hist_range[1], 500) # x-values for KDE plots
    #colors = ['#1f77b4', '#ff7f0e'] 
    colors = ['#daa520', '#3cb371']
    label_bo = ["Experimental Structure", "Predicted Structure"]
    for i, (label, omega_angles) in enumerate(omega_data_dict.items()):
        if not omega_angles:
            print(f"No omega angles for {label}. Skipping plot for this dataset.")
            continue

        # Angle Transformation: Shift angles from [-180, -90) to [180, 270)
        # to center the trans peak and align cis at 0.
        adjusted_omega_angles = [angle + 360 if angle < -90 else angle for angle in omega_angles]

        if len(adjusted_omega_angles) > 1: # KDE requires more than one point
            kde = gaussian_kde(adjusted_omega_angles)
            y_kde = kde(x_kde)
            
            # Plot the raw density for comparison
            # Using linewidth=2.5 and alpha=0.9 for lines, inspired by bar plot width/alpha
            ax1.plot(x_kde, y_kde, color=colors[i % len(colors)], linewidth=2.5, alpha=0.9, linestyle='-', label=f'{label_bo[i]}')
        else:
            print(f"Not enough data points for KDE for {label}.")

    # --- Aesthetics and Labels ---
    ax1.set_xlabel("Omega (ω) Angle [º]", fontsize=14)
    ax1.set_ylabel("Probability density", fontsize=14) # Y-axis represents probability density
    
    # no title as there is already description in the work!
    #ax1.set_title(f"Distribution of Omega Angles for {title_suffix}", fontsize=14, weight='bold')
    
    # --- Set X-axis limits and ticks ---
    ax1.set_xlim(hist_range)
    ax1.set_xticks(np.arange(hist_range[0], hist_range[1] + 1, 20))

    # --- Vertical Lines for Cis/Trans (subtler for comparison plots) ---
    ax1.axvline(180, color='gray',linestyle='--', linewidth=1.5, label='Trans conformation (180°)')
    
    # Legend placement similar to triple_ploter
    fig.legend(loc="upper right", bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes, fontsize=14 )
    
    plt.tight_layout() # Adjust plot to prevent labels from overlapping
    
    # --- Save Plot ---
    # Sanitize title_suffix for filename to avoid issues with special characters
    sanitized_suffix = title_suffix.replace(' ', '_').replace('(', '').replace(')', '').replace('/', '_')
    output_pdf_path = os.path.join(output_directory, f"omega_comparison_plot_{sanitized_suffix}.pdf")
    plt.savefig(output_pdf_path)
    print(f"Comparison plot saved to {output_pdf_path}")
    plt.show()
    plt.close() # Close the plot to free memory

# --- Main Execution Block (No change) ---
if __name__ == "__main__":
    # Define your protein file names
    main_protein_name = "9cah" 
    protein1_name = main_protein_name 
    protein2_name = main_protein_name + "_pred" 

    # Extract omega angles for each protein
    omega_angles_prot1 = extract_omega_angles(protein1_name)
    omega_angles_prot2 = extract_omega_angles(protein2_name)

    # --- Prepare directory and file for saving statistics (in 'omegas' folder) ---
    output_directory = "omegas"
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    stats_file_name = f"omega_stats_{protein1_name}_vs_{protein2_name}.txt"
    stats_file_path = os.path.join(output_directory, stats_file_name)

    # --- Calculate and save statistics for each protein ---
    print("\n--- Saving Omega Angle Statistics ---")
    with open(stats_file_path, 'w') as f:
        # For protein 1
        if omega_angles_prot1:
            adjusted_omega_angles_prot1 = [angle + 360 if angle < -90 else angle for angle in omega_angles_prot1]
            mean_prot1 = np.mean(adjusted_omega_angles_prot1)
            std_prot1 = np.std(adjusted_omega_angles_prot1)
            
            f.write(f"{protein1_name}:\n")
            f.write(f"Mean Omega Angle: {mean_prot1:.2f} degrees\n")
            f.write(f"Standard Deviation: {std_prot1:.2f} degrees\n\n")
        else:
            f.write(f"No angles for {protein1_name} to calculate statistics.\n\n")

        # For protein 2
        if omega_angles_prot2:
            adjusted_omega_angles_prot2 = [angle + 360 if angle < -90 else angle for angle in omega_angles_prot2]
            mean_prot2 = np.mean(adjusted_omega_angles_prot2)
            std_prot2 = np.std(adjusted_omega_angles_prot2)
            
            f.write(f"{protein2_name}:\n")
            f.write(f"  Mean Omega Angle: {mean_prot2:.2f} degrees\n")
            f.write(f"  Standard Deviation: {std_prot2:.2f} degrees\n")
        else:
            f.write(f"No angles for {protein2_name} to calculate statistics.\n")
    
    print(f"Omega angle statistics saved to: {stats_file_path}")

    # Store angles in a dictionary for easy plotting
    all_omega_data = {
        protein1_name: omega_angles_prot1,
        protein2_name: omega_angles_prot2
    }
    print("Cis residues experimental")
    for i in range(len( all_omega_data[protein1_name])):
        if (all_omega_data[protein1_name][i]<90 and all_omega_data[protein1_name][i]>-30):
            print(i, all_omega_data[protein1_name][i])
    print("Cis residues predicted")
    for i in range(len( all_omega_data[protein2_name])):
        if (all_omega_data[protein2_name][i]<90 and all_omega_data[protein2_name][i]>-30):
            print(i, all_omega_data[protein2_name][i])
    #fixing too small angles
    for i in range(len( all_omega_data[protein1_name])):
        if (all_omega_data[protein1_name][i]<-90):
            all_omega_data[protein1_name][i] = all_omega_data[protein1_name][i] +360
    for i in range(len( all_omega_data[protein2_name])):
        if (all_omega_data[protein2_name][i]<-90):
            all_omega_data[protein2_name][i] = all_omega_data[protein2_name][i] +360
    # Plot the distributions for comparison
    plot_omega_distributions(all_omega_data, title_suffix=f"{protein1_name} vs {protein2_name}", output_directory=output_directory)
```
