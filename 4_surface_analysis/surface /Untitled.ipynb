```python
from Bio.PDB import PDBParser

parser = PDBParser()
structure = parser.get_structure("model", "1bl8_model.pdb")

embeded_area = 0
for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                print(atom.get_bfactor())
                print(atom.get_name(), atom.coord)
```
