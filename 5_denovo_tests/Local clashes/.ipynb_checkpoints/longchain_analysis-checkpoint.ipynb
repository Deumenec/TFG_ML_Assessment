```python
#Disorder comparison plots. In this plot, the disorder associated with a given structure is compared with the pLDDT of an alphafold model, showing how there is no good correlation
import json
import matplotlib.pyplot as plt

with open('prediction_data_7F60.json', 'r') as f:
    data1 = json.load(f)  # This loads the file content into a Python dictionary

with open('AIUPred_7F60.txt', 'r') as f:
    data = f.readlines()
    dades = []
    for dat in range(9, len(data)-1):
        dades.append(data[dat].split("\t"))
len_aminoacids = {
    "A": 5,  # ALA
    "R": 11, # ARG
    "N": 8,  # ASN
    "D": 8,  # ASP
    "C": 6,  # CYS
    "Q": 9,  # GLN
    "E": 9,  # GLU
    "G": 4,  # GLY
    "H": 10, # HIS
    "I": 8,  # ILE
    "L": 8,  # LEU
    "K": 9,  # LYS
    "M": 8,  # MET
    "F": 11, # PHE
    "P": 7,  # PRO
    "S": 6,  # SER
    "T": 7,  # THR
    "W": 14, # TRP
    "Y": 12, # TYR
    "V": 7   # VAL
}
    
```

```python
print("Keys from the dictionary are: "+ str(data1.keys()))
atom_disorder=[]
id_num =[0]
for dada in dades:
    for i in range(len_aminoacids[dada[1]]):
        atom_disorder.append(float(dada[2])*100)
        id_num.append(int(dada[0]))
chain = data1["atom_chain_ids"]
plddt = data1["atom_plddts"]
#print(plddt)
analized_chain= []
analized_chain_name = "E"
for i in range(len(chain)):
    if chain[i] == analized_chain_name: analized_chain.append(plddt[i])
plt.plot(analized_chain, color='#daa520')
plt.plot(atom_disorder, color='#3cb371')
```
