```python
#Disorder comparison plots. In this plot, the disorder associated with a given structure is compared with the pLDDT of an alphafold model, showing how there is no good correlation
import json
import matplotlib.pyplot as plt

with open('prediction_data_7F60.json', 'r') as f:
    data1 = json.load(f)  # This loads the file content into a Python dictionary

with open('prediction_data_alone.json', 'r') as f:
    data2 = json.load(f)  # This loads the file content into a Python dictionary

with open('AIUPred_7F60.txt', 'r') as f:
    data = f.readlines()
    dades = []
    for dat in range(9, len(data)-1):
        dades.append(data[dat].split("\t"))
len_aminoacids = {
    "A": 5,  # ALA
    "R": 11, # ARG
    "N": 8,  # ASN
    "D": 8,  # ASP
    "C": 6,  # CYS
    "Q": 9,  # GLN
    "E": 9,  # GLU
    "G": 4,  # GLY
    "H": 10, # HIS
    "I": 8,  # ILE
    "L": 8,  # LEU
    "K": 9,  # LYS
    "M": 8,  # MET
    "F": 11, # PHE
    "P": 7,  # PRO
    "S": 6,  # SER
    "T": 7,  # THR
    "W": 14, # TRP
    "Y": 12, # TYR
    "V": 7   # VAL
}
    
```

```python
#Per atom confidence plot
print("Keys from the dictionary are: "+ str(data1.keys()))
atom_disorder=[]
id_num =[]
for dada in dades:
    for i in range(len_aminoacids[dada[1]]):
        atom_disorder.append(float(dada[2])*100)
        id_num.append(int(dada[0]))
chain = data1["atom_chain_ids"]
plddt = data1["atom_plddts"]
#print(plddt)
analized_chain= []
analized_chain_name = "E"
for i in range(len(chain)):
    if chain[i] == analized_chain_name: analized_chain.append(plddt[i])
drawing = plt.plot(analized_chain, color='#daa520', label= "Predicted LDDT")
drawing += plt.plot(atom_disorder, color='#3cb371', label ="Disorder probability")
plt.legend(loc="upper right")
plt.xlabel("Atom Index")
plt.ylabel("pLDDT | Disorder probability (%)")
plt.savefig("pLDDTvsOrder.pdf")
```

```python
#Per residue confidence plot
print("Keys from the dictionary are: "+ str(data1.keys()))
atom_disorder=[]
id_num =[]
for dada in dades:
    atom_disorder.append(float(dada[2])*100)
    for i in range(len_aminoacids[dada[1]]):
        id_num.append((dada[1]))
chain = data1["atom_chain_ids"]
plddt = data1["atom_plddts"]
chain2= data2["atom_chain_ids"]
plddt2= data2["atom_plddts"]
#print(plddt)
analized_chain1= []
analized_chain2= plddt2
analized_chain_name1 = "E"
for i in range(len(chain)):
    if chain[i] == analized_chain_name1: analized_chain1.append(plddt[i])


residue_pLDDTalone = []
residue_pLDDT =[]
subatom = 0
subaverage1 = 0
subaverage2 = 0
#Weighted average to predict average pLDDT on all residues
for i in range(len(id_num)):
    if subatom ==0:
        subatom = len_aminoacids[id_num[i]]
        subaverage1 =0
        subaverage2 =0
    subaverage1 += analized_chain1[i]
    subaverage2 += analized_chain2[i]
    subatom += -1
    if subatom ==0:
        residue_pLDDT.append(subaverage1/len_aminoacids[id_num[i]])
        residue_pLDDTalone.append(subaverage2/len_aminoacids[id_num[i]])



plt.figure(figsize=(11, 5))
plt.tight_layout()
#plt.plot(residue_pLDDT, color= "#6c8ebf", label= "Predicted LDDT") #This plots the pLDDT of the predicted structure
plt.plot(atom_disorder, color='#3cb371', label ="Disorder probability")
plt.plot(residue_pLDDTalone, color='#daa520', label= "Predicted LDDT")

plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Residue Index", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("pLDDT | Disorder probability (%)", fontsize=14)
plt.autoscale(enable=True, axis='x', tight=True)
plt.autoscale(enable=True, axis='y', tight=True)
plt.axhspan(ymin =0, ymax=119, color= "white")
plt.savefig("pLDDTvsOrder.pdf")
```
